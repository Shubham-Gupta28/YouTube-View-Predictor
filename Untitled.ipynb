```python
def get_channel_uploads_playlist_id(channel_id):
    """
    Get the uploads playlist ID for a channel
    This is more efficient than searching
    """
    try:
        request = youtube.channels().list(
            part='contentDetails',
            id=channel_id
        )
        response = request.execute()
        
        if 'items' in response and len(response['items']) > 0:
            uploads_playlist_id = response['items'][0]['contentDetails']['relatedPlaylists']['uploads']
            return uploads_playlist_id
        else:
            return None
    except HttpError as e:
        print(f"Error getting playlist: {e}")
        return None

def get_videos_from_playlist(playlist_id, max_results=1000):
    """
    Fetch all videos from a playlist (uploaded videos)
    """
    video_ids = []
    next_page_token = None
    total_fetched = 0
    
    print(f"Attempting to fetch up to {max_results} videos...")
    
    while total_fetched < max_results:
        try:
            request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId=playlist_id,
                maxResults=min(50, max_results - total_fetched),
                pageToken=next_page_token
            )
            response = request.execute()
            
            # Extract video IDs
            for item in response['items']:
                video_ids.append(item['contentDetails']['videoId'])
                total_fetched += 1
            
            next_page_token = response.get('nextPageToken')
            
            if not next_page_token:
                break
                
            print(f"Fetched {total_fetched} videos so far...")
            
        except HttpError as e:
            print(f"Error: {e}")
            break
    
    return video_ids

def get_video_details_batch(video_ids, batch_size=50):
    """
    Get detailed metadata for videos in batches
    """
    all_video_data = []
    total_batches = (len(video_ids) + batch_size - 1) // batch_size
    
    print(f"Fetching details for {len(video_ids)} videos in {total_batches} batches...")
    
    for i in range(0, len(video_ids), batch_size):
        batch_ids = video_ids[i:i+batch_size]
        batch_num = (i // batch_size) + 1
        try:
            request = youtube.videos().list(
                part='snippet,statistics,contentDetails',
                id=','.join(batch_ids)
            )
            response = request.execute()
            
            for video in response['items']:
                # Parse duration from ISO 8601 format
                duration_str = video['contentDetails']['duration']
                
                video_data = {
                    'video_id': video['id'],
                    'title': video['snippet']['title'],
                    'description': video['snippet']['description'],
                    'published_at': video['snippet']['publishedAt'],
                    'channel_title': video['snippet']['channelTitle'],
                    'tags': video['snippet'].get('tags', []),
                    'category_id': video['snippet'].get('categoryId', ''),
                    'duration': duration_str,
                    'view_count': int(video['statistics'].get('viewCount', 0)),
                    'like_count': int(video['statistics'].get('likeCount', 0)),
                    'comment_count': int(video['statistics'].get('commentCount', 0)),
                }
                all_video_data.append(video_data)
            
            print(f"Batch {batch_num}/{total_batches}: Fetched details for {len(all_video_data)} videos...")
            
        except HttpError as e:
            print(f"Error fetching video details in batch {batch_num}: {e}")
            continue
    
    return all_video_data

# Collect videos from all channels
all_channel_data = {}
videos_per_channel = 900

print("Starting Video Collection")


for channel_name, channel_id in CHANNELS.items():
    print(f"\n{'='*50}")
    print(f"Processing: {channel_name}")
    print(f"Channel ID: {channel_id}")
    
    # Get uploads playlist ID
    playlist_id = get_channel_uploads_playlist_id(channel_id)
    
    if playlist_id:
        print(f"Uploads playlist ID: {playlist_id}")
        
        # Get all video IDs from playlist
        video_ids = get_videos_from_playlist(playlist_id, max_results=videos_per_channel)
        print(f"Total videos found: {len(video_ids)}")
        
        if video_ids:
            # Get detailed metadata
            video_data = get_video_details_batch(video_ids)
            all_channel_data[channel_name] = video_data
            print(f"SUCCESS: Fetched metadata for {len(video_data)} videos from {channel_name}")
        else:
            print(f"No videos found for {channel_name}")
    else:
        print(f"Could not get uploads playlist for {channel_name}")

# Create summary
print("SUMMARY")
total_collected = 0
for channel_name, videos in all_channel_data.items():
    print(f"{channel_name:20} : {len(videos):4} videos")
    total_collected += len(videos)

print(f"TOTAL VIDEOS COLLECTED: {total_collected}")

# Combine all data into one dataframe
all_videos = []
for channel_name, videos in all_channel_data.items():
    for video in videos:
        video['channel_name'] = channel_name
        all_videos.append(video)

df_raw = pd.DataFrame(all_videos)
print(f"\nDataFrame created with {len(df_raw)} rows")

# Display sample
print(f"\nSample data (first 5 videos):")
display(df_raw.head())

# Save raw data to CSV for backup
df_raw.to_csv('youtube_videos_raw.csv', index=False)
print(f"\nRaw data saved to 'youtube_videos_raw.csv'")

# Display data types and basic info
print("\nDataFrame Info:")
print(df_raw.info())

# Check for missing values
print("\nMissing Values:")
print(df_raw.isnull().sum())

# Distribution per channel
print("\nVideos per channel:")
print(df_raw['channel_name'].value_counts())

# Additional statistics
print("\nBasic Statistics:")
print(f"Total videos: {len(df_raw)}")
print(f"Total channels: {df_raw['channel_name'].nunique()}")
print(f"Date range: {df_raw['published_at'].min()} to {df_raw['published_at'].max()}")
print(f"Average views: {df_raw['view_count'].mean():,.0f}")
print(f"Average likes: {df_raw['like_count'].mean():,.0f}")
print(f"Average comments: {df_raw['comment_count'].mean():,.0f}")
```
