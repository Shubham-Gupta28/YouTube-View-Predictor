```python
# Basic temporal features
df['publish_year'] = df['published_at'].dt.year
df['publish_month'] = df['published_at'].dt.month
df['publish_day'] = df['published_at'].dt.day
df['publish_hour'] = df['published_at'].dt.hour
df['publish_minute'] = df['published_at'].dt.minute
df['publish_dayofweek'] = df['published_at'].dt.dayofweek
df['publish_quarter'] = df['published_at'].dt.quarter
df['publish_day_name'] = df['published_at'].dt.day_name()

# Weekend/Weekday flags
df['is_weekend'] = df['publish_dayofweek'].isin([5, 6]).astype(int)
df['is_weekday'] = (df['publish_dayofweek'] < 5).astype(int)

# Peak time flags (based on EDA insights)
df['is_saturday_5pm'] = ((df['publish_dayofweek'] == 5) & (df['publish_hour'] == 17)).astype(int)
df['is_peak_hour'] = df['publish_hour'].isin([16, 17, 19, 20, 21]).astype(int)

# Time category
def get_time_category(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['time_category'] = df['publish_hour'].apply(get_time_category)
```

```python
df['tag_count'] = df['tags'].apply(len)
df['has_tags'] = (df['tag_count'] > 0).astype(int)
```

```python
def parse_duration(duration_str):
    duration = isodate.parse_duration(duration_str)
    return duration.total_seconds()
df['duration_seconds'] = df['duration'].apply(parse_duration)
df['duration_minutes'] = df['duration_seconds'] / 60

def get_duration_category(minutes):
    if minutes < 1:
        return 'Very Short'
    elif minutes < 5:
        return 'Short'
    elif minutes < 10:
        return 'Medium'
    elif minutes < 20:
        return 'Long'
    else:
        return 'Very Long'

df['duration_category'] = df['duration_minutes'].apply(get_duration_category)
df['is_short'] = (df['duration_minutes'] < 10).astype(int)
df['is_long'] = (df['duration_minutes'] > 20).astype(int)
```

```python
# Days Since Last Upload
df = df.sort_values(['channel_name', 'published_at'])
df['days_since_last_upload'] = df.groupby('channel_name')['published_at'].diff().dt.days
df['days_since_last_upload'] = df['days_since_last_upload'].fillna(0)
```

```python
# Regression target
df['views_7day'] = df['view_count']

# Classification targets
df['will_cross_100k'] = (df['view_count'] > 100000).astype(int)
df['will_cross_1M'] = (df['view_count'] > 1000000).astype(int)
df['will_cross_10M'] = (df['view_count'] > 10000000).astype(int)
```

```python
#Channel-Specific Features
channel_avg_views = df.groupby('channel_name')['view_count'].mean().to_dict()
channel_avg_likes = df.groupby('channel_name')['like_count'].mean().to_dict()
channel_upload_freq = df.groupby('channel_name')['published_at'].count().to_dict()

df['channel_avg_views'] = df['channel_name'].map(channel_avg_views)
df['channel_avg_likes'] = df['channel_name'].map(channel_avg_likes)
df['channel_upload_freq'] = df['channel_name'].map(channel_upload_freq)
```

```python
# Encode categorical variables
le_channel = LabelEncoder()
le_time = LabelEncoder()
le_duration = LabelEncoder()

df['channel_encoded'] = le_channel.fit_transform(df['channel_name'])
df['time_category_encoded'] = le_time.fit_transform(df['time_category'])
df['duration_category_encoded'] = le_duration.fit_transform(df['duration_category'])
```

```python
feature_cols = ['publish_year', 'publish_month', 'publish_hour', 'publish_dayofweek',
    'is_weekend', 'is_weekday', 'is_saturday_5pm', 'is_peak_hour',
    'days_since_last_upload', 'title_length', 'tag_count', 'has_tags',
    'duration_minutes', 'is_short', 'is_long','channel_avg_views', 'channel_avg_likes', 'channel_upload_freq',
    'channel_encoded', 'time_category_encoded', 'duration_category_encoded'
]

X = df[feature_cols]
y_reg = df['views_7day']
y_clf_100k = df['will_cross_100k']  # Classification: 100K views
y_clf_1M = df['will_cross_1M']      # Classification: 1M views
y_clf_10M = df['will_cross_10M']    # Classification: 10M views
```

```python
# Train-Test Split and Scaling
X_train, X_test, y_reg_train, y_reg_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42, stratify=df['channel_name']
)

# Get indices for classification targets
train_indices = X_train.index
test_indices = X_test.index

y_clf_100k_train = y_clf_100k.loc[train_indices]
y_clf_100k_test = y_clf_100k.loc[test_indices]

y_clf_1M_train = y_clf_1M.loc[train_indices]
y_clf_1M_test = y_clf_1M.loc[test_indices]

y_clf_10M_train = y_clf_10M.loc[train_indices]
y_clf_10M_test = y_clf_10M.loc[test_indices]

print(f"Train size: {len(X_train)} videos")
print(f"Test size: {len(X_test)} videos")

print("\nTrain split by channel:")
for channel in df['channel_name'].unique():
    train_count = (df.loc[train_indices, 'channel_name'] == channel).sum()
    test_count = (df.loc[test_indices, 'channel_name'] == channel).sum()
    train_pct = train_count / (train_count + test_count) * 100
    print(f"  {channel}: Train {train_count} ({train_pct:.0f}%), Test {test_count}")

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("COMPARING MULTIPLE MODELS")
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(max_depth=10, random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, random_state=42, verbosity=0),
    'KNN': KNeighborsRegressor(n_neighbors=5)
}

# Store results
results = []

print("Training and evaluating models...")


for name, model in models.items():
    start_time = time.time()
    
    # Train model
    model.fit(X_train_scaled, y_reg_train)
    
    # Predict
    y_pred = model.predict(X_test_scaled)
    
    # Calculate metrics
    mse = mean_squared_error(y_reg_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_reg_test, y_pred)
    r2 = r2_score(y_reg_test, y_pred)
    results.append({
        'Model': name,
        'RMSE': rmse,
        'MAE': mae,
        'R2': r2
    })
    
    print(f"{name:20} R2: {r2:.4f}, RMSE: {rmse:,.0f}")

# Create results dataframe
results_df = pd.DataFrame(results)
results_df = results_df.round(4)

print("\n")
print("MODEL COMPARISON SUMMARY")
print(results_df.to_string(index=False))

# Find best model
best_model = results_df.loc[results_df['R2'].idxmax()]
print(f"\nBest Model: {best_model['Model']}")
print(f"R2 Score: {best_model['R2']:.4f}")
print(f"RMSE: {best_model['RMSE']:,.0f} views")
```

```python
# Compare Classification Models for All Thresholds

print("COMPARING CLASSIFICATION MODELS FOR ALL THRESHOLDS")

# List of classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(max_depth=10, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(n_estimators=100, random_state=42, verbosity=0),
    'KNN': KNeighborsClassifier(n_neighbors=5)
}

# Define thresholds
thresholds = {
    '100K': (y_clf_100k_train, y_clf_100k_test, '100K Views'),
    '1M': (y_clf_1M_train, y_clf_1M_test, '1M Views'),
    '10M': (y_clf_10M_train, y_clf_10M_test, '10M Views')
}

# Store all results
all_results = []

for threshold_key, (y_train, y_test, threshold_name) in thresholds.items():

    print(f"\nTHRESHOLD: {threshold_name}")

    
    print("\nTraining and evaluating classifiers...")

    
    for clf_name, model in classifiers.items():
        start_time = time.time()
        
        # Train model
        model.fit(X_train_scaled, y_train)
        
        # Predict
        y_pred = model.predict(X_test_scaled)
        
        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)
        
        train_time = time.time() - start_time
        
        all_results.append({
            'Threshold': threshold_name,
            'Model': clf_name,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1,
            'Train Time (s)': train_time
        })
        
        print(f"{clf_name:20} Accuracy: {accuracy:.4f}, F1: {f1:.4f}, Time: {train_time:.2f}s")

# Create results dataframe
results_df = pd.DataFrame(all_results)
results_df = results_df.round(4)

# Display results by threshold
for threshold_key, (_, _, threshold_name) in thresholds.items():
    print("\n")
    print(f"RESULTS FOR {threshold_name.upper()}")

    
    threshold_results = results_df[results_df['Threshold'] == threshold_name]
    print(threshold_results.to_string(index=False))
    
    # Find best model for this threshold
    best_idx = threshold_results['Accuracy'].idxmax()
    best_row = threshold_results.loc[best_idx]
    print(f"\nBest Model for {threshold_name}: {best_row['Model']}")
    print(f"  Accuracy: {best_row['Accuracy']:.4f}")
    print(f"  F1 Score: {best_row['F1 Score']:.4f}")

# Overall comparison
print("\n")
print("OVERALL MODEL COMPARISON BY THRESHOLD")

# Pivot table for easy comparison
pivot_acc = results_df.pivot(index='Model', columns='Threshold', values='Accuracy')
pivot_f1 = results_df.pivot(index='Model', columns='Threshold', values='F1 Score')

print("\nAccuracy by Threshold:")
print(pivot_acc.round(4))

print("\nF1 Score by Threshold:")
print(pivot_f1.round(4))

# Find best overall model
best_overall = results_df.loc[results_df.groupby('Threshold')['Accuracy'].idxmax()]

print("\nBEST MODEL FOR EACH THRESHOLD")

print(best_overall[['Threshold', 'Model', 'Accuracy', 'F1 Score']].to_string(index=False))

for threshold_key, (y_train, y_test, threshold_name) in thresholds.items():
    # Find best model for this threshold
    threshold_results = results_df[results_df['Threshold'] == threshold_name]
    best_idx = threshold_results['Accuracy'].idxmax()
    best_row = threshold_results.loc[best_idx]
    best_model_name = best_row['Model']
    best_model = classifiers[best_model_name]
    
    # Retrain best model on full training data
    best_model.fit(X_train_scaled, y_train)
```
